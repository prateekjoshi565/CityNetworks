```python
import networkx as nx
import osmnx as ox
import matplotlib.pyplot as plt
```

```python
# Bengaluru
G = ox.graph_from_point((12.962714, 77.589869), dist=2000, network_type="drive")
fig, ax = ox.plot_graph(G, node_size=10, edge_color = "#BF3EFF", node_color= "#EE1289", edge_linewidth=1, show=False, close=False)
ax.set_title('Bengaluru')
plt.show()
```
